```python
import pandas as pd
import numpy as np

#importing dataset into python
shots_2020 = pd.read_csv("shots_2020.csv")
shots_2021 = pd.read_csv("shots_2021.csv")

#choosing specific columns for dataframe
df_2020 = shots_2020[['shotAngleAdjusted', 'shotDistance', 'shotType', 'shotRush', 'homeTeamGoals', 'awayTeamGoals', 'offWing', 'shotRebound', 'defendingTeamAverageTimeOnIce', 'shootingTeamAverageTimeOnIce', 'isHomeTeam']]
df_2021 = shots_2021[['shotAngleAdjusted', 'shotDistance', 'shotType', 'shotRush', 'homeTeamGoals', 'awayTeamGoals', 'offWing', 'shotRebound', 'defendingTeamAverageTimeOnIce', 'shootingTeamAverageTimeOnIce', 'isHomeTeam']]

#combining dataframes so all the data is in one dataframe
df = pd.concat([df_2020, df_2021])

#changing shotType to numerics
df['shotType'] = df['shotType'].replace(['WRIST','SNAP','SLAP','TIP','BACK','DEFL'],[0,1,2,3,4,5])

#checking to see if shotType has changed
df.head()
```
